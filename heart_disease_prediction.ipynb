```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test  ,y_train,y_test=train_test_split(x200,Y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
n_estimators=[20,60,100,120,300,400,450]
max_depth=[10,20,40,None]
max_features=[0.2,0.4,0.8,1.0]
max_samples=[0.2,0.4,0.8,1.0]
param_grid={'n_estimators':n_estimators,'max_depth':max_depth,'max_features':max_features,'max_samples':max_samples}
clf=GridSearchCV(RandomForestClassifier(),param_grid,cv=5)



clf.fit(x_train, y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
