```python
def aryan(x):
  tokens=x.split('-')
  if len(tokens)==2:
    return(float(tokens[0])+float(tokens[1])/2)
  try:
    return float(x)
  except:
    return None
```

```python
x4=x.copy()
x4['cgpa']=x4['cgpa'].apply(aryan)
x4.head()
```
